（12）用数据透视表，计算每个类型的酒店的评分人数总数量。

```python
pd.pivot_table(df,index=['类型'],values=['评分人数'],aggfunc=np.sum)
```

（13）用数据透视表，计算每个类型的酒店价格的均值和标准差

```python
pd.pivot_table(df,index=['类型'],values=['价格'],aggfunc=[np.mean,np.std])
```

（14）用数据透视表，计算每个地区酒店价格和评分的最大值和最小值

```python
pd.pivot_table(df,index=['地区'],values=['价格','评分'],aggfunc=[np.max,np.min])
```

（15）用数据透视表，计算每个地区和类型的酒店的评分的均值和标准差

```python
pd.pivot_table(df,index=['地区','类型'],values=['评分'],aggfunc=[np.mean,np.std])
```
