（1）读取数据。读取之前作业保存的“酒店数据1.xlsx”

```python
df = pd.read_excel("酒店数据1.xlsx",index_col=0)
df.head()
```

（2）将“类型”和“名字”设置为层次化索引，并交换索引的位置。然后将层次化索引取消。

```python
df = df.set_index(["类型","名字"])
df
```

```python
df = df.swaplevel('类型','名字')
df
```

```python
df = df.reset_index()
df
```

（3）将数据集转置，获取转制后的index和columns。

```python
temp = df.transpose()
temp.index
```

（4）用Groupby方法来计算每个地区的评分人数的总和以及均值。

```python
df['评分人数'].groupby(by = df['地区']).sum()
```

```python
df['评分人数'].groupby(by = df['地区']).mean()
```

（5）用Grouby方法计算每个类型的平均价格，最高价和最低价。

```python
df['价格'].groupby(by = df['类型']).mean()
```

```python
df['价格'].groupby(by = df['类型']).max()
```

```python
df['价格'].groupby(by = df['类型']).min()
```

（6）数据离散化，按照价格将酒店分为3个等级,0-500为C，500-1000为B,大于1000为A，列名设置为“价格等级”。

```python
df['价格等级'] = pd.cut(df['价格'],bins=[0,500,1000,df['价格'].max()],labels=['C','B','A'])
df.head()
```

（7）获取评分均值最高和最低的地区的数据，分别使用append和concat方法将获取的两个数据集合并。

```python
df.groupby(by=df['地区']).mean().sort_values(by='评分')
```

```python
temp1 = df[df['地区'] == '葵青']
temp2 = df[df['地区'] == '屯门']
temp1.append(temp2)
```

```python
pd.concat([temp1,temp2],axis=0)
```

（8）数据离散化，按照评分人数将酒店平均分为3个等级，三个等级的酒店数量尽量保持一致。评分人数最多的为A，最少的为C。列名设置为“热门等级”。

```python
bins = np.percentile(df['评分人数'],[0,33,66,100])
df['热门等级'] = pd.cut(df['评分人数'], bins=bins, labels=['C','B','A'])
df.head()
```

（9）选出评分人数为A，价格也为A的酒店数据，计算其平均评分。

```python
df[(df['热门等级'] == 'A') & (df['价格等级'] == 'A')].head()
```

```python
df[(df['热门等级'] == 'A') & (df['价格等级'] == 'A')]['评分'].mean()
```

（10）取价格最高的5个酒店的数据，使用stack和unstack方法实现dataframe和Series之间的转换。

```python
df.sort_values(by='价格',ascending=False)[:5].stack()
df.sort_values(by='价格',ascending=False)[:5].stack().unstack()
```

（11）纵向拆分数据集，分为df1和df2，df1包含名字，类型，城市，地区，df2包含名字，地点，评分，评分人数，价格，价格等级，热门等级。

```python
df1 = df[['名字','类型','城市','地区']]
df2 = df[['名字','地点','评分','评分人数','价格','价格等级','热门等级']]
```

（12）将df2按照价格进行排序，重新设置df2的索引。索引值等于价格排名。

```python
df2 = df2.sort_values(by='价格',ascending=False)
df2.index = range(len(df2))
df2
```

（13）使用merge方法将df1和df2合并。

```python
df3 = pd.merge(df1, df2, how="inner", on='名字')
df3
```

（14）将合并后的数据集保存数据到“酒店数据2.xlsx”。

```python
df3.to_excel("酒店数据2.xlsx")
```
