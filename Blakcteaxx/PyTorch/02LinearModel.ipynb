>### Linear Regression
- Evaluate Model Error:\
    - Training Loss(Error): targeted at 样本
    - Cost:MAE,MSE; Targeted at the training set
- Compute Loss:\
    Closed to zero is the Best
- How to find the Loss min?

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

def linearmodel(x):
    return w * x

def loss(x, y):
    y_pred = linearmodel(x)
    return (y_pred - y) ** 2

mse_list = []

for w in np.arange(0,4.1,0.1):
    mse = 0
    for i in range(len(x_data)):
        mse = mse + loss(x_data[i], y_data[i])
    mse_list.append(mse)

plt.plot(np.arange(0,4.1,0.1),mse_list,"b")
plt.xlabel(r'$w$')
plt.ylabel(r"$MSE$")
plt.show()
```
