```python
temp = {
    "state":"e",
    "year":1995,
    "pop":10
}
line = pd.Series(temp)
line.name = "E"
result = result.append(line)
result
```

```python
result = result.drop(["E"],axis=0)
result
```

（6）获取前‘state’和‘year’的数据。

```python
result[["state","year"]]
```

## （2）数据操作

（1）读取香港酒店数据。

```python
df = pd.read_excel("香港酒店数据.xlsx")
df.head()
```

（2）按照数据的内容，重新设置数据的索引，重新设置列名称为'名字','类型','城市','地区','地点','评分','评分人数','价格'。

```python
columns = ["ID",'名字','类型','城市','地区','地点','评分','评分人数','价格']
df.columns = columns
df.head()
```

（3）查看所有类型为“浪漫情侣”的酒店

```python
df[df["类型"] == "浪漫情侣"]
```

（4）查看所有类型为“浪漫情侣”，地区在湾仔的酒店

```python
df[(df["类型"] == "浪漫情侣") & (df["地区"] == "湾仔")]
```

（5）查看所有地址在观塘或者油尖旺，评分大于4的酒店

```python
df[(df["评分"] > 4) & (df["地区"] == "观塘") | (df["地区"] == "油尖旺")]
```

（6）查看类型缺失的数据

```python
df[df['类型'].isnull()]
```

（7）用“其他”填充类型和地区

```python
df['类型'].fillna("其他",inplace=True)
df['地区'].fillna("其它",inplace=True)
df
```

（8）用评分均值填充缺失值

```python
df["评分"].fillna(np.mean(df['评分']),inplace=True)
df
```

（9）删除价格和评分人数的缺失值

```python
df.dropna(axis=0,inplace=True,subset=["价格","评分人数"],how="any")
```
