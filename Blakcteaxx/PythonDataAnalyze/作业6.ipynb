```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
x = np.arange(-5,3.1,0.1)
y = x * x + 2 * x + 1
plt.plot(x, y, "b--")
plt.show()
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
x = np.linspace(-np.pi,np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

# plt.figure(figsize=(16,18))

plt.subplot(211)
plt.plot(x,y1,linewidth='2.5',color='b')

plt.subplot(212)
plt.plot(x,y2,linewidth='2.5',color='g')

plt.plot()
```

（3）读取上次作业保存的酒店数据，画出每个地区酒店数量的柱状图，柱状颜色为红色

```python
df = pd.read_excel("酒店数据2.xlsx",index_col=0)
df.head()
```

```python
import warnings
warnings.filterwarnings("ignore")

plt.rcParams["font.sans-serif"] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

data = df["地区"].value_counts()
x = data.index
y = data.values

plt.bar(x,y,color='r')
plt.title("各地区酒店数量",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

（4）画出每个价格等级酒店数量的柱状图。

```python
data = df['价格等级'].value_counts()
x = data.index
y = data.values

plt.bar(x,y,color='b')
plt.title("各价格等级酒店数量",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

（5）画出各个价格等级占比的饼图。

```python
data = df['价格等级'].value_counts()
x = data.index
y = data.values

plt.pie(y,labels=x,startangle=90,autopct="%.1f %%")
plt.title("各价格等级酒店数量")
plt.show()
```

（6）画出酒店评分的直方图。

```python
plt.figure(figsize = (10,6))
plt.hist(df["评分"], bins = 50, edgecolor = 'k',alpha = 0.5)
plt.show()
```

（7）画出每个热门等级酒店评分均值的柱状图。（按照评分均值从小到大排序。）

```python
data = df["评分"].groupby(df["热门等级"]).mean().sort_values(ascending=True)

x = data.index
y = data.values

plt.bar(x,y,color='b')
plt.title("每个热门等级酒店评分均值",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```
