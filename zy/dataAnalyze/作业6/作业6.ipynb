```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
def f(x):
    return x*x+2*x+1

x = np.arange(-5,3,0.1)
plt.plot(x,f(x))
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
def f1(x):
    return np.sin(x)
def f2(x):
    return np.cos(x)
x = np.arange(-3.14,3.14,0.1)
plt.subplot(211)
t = plt.plot(x,f1(x),linewidth = 2.5)
plt.subplot(212)
plt.plot(x,f2(x),linewidth = 2.5)
plt.show()
```

（3）读取上次作业保存的酒店数据，画出每个地区酒店数量的柱状图，柱状颜色为红色

```python
plt.rcParams["font.sans-serif"] = ["SimHei"] #解决中文字符乱码的问题
plt.rcParams["axes.unicode_minus"] = False #正常显示负号
```

```python
df = pd.read_excel("酒店数据2.xlsx")
l = df["地区"].value_counts()
x,y = l.index,l.values
plt.bar(x,y,color = 'r')
plt.xticks(rotation = 90) #标签转90度
plt.show()
```

（4）画出每个价格等级酒店数量的柱状图。

```python
l = df["等级"].value_counts()
x,y = l.index,l.values
plt.bar(x,y)
```

（5）画出各个价格等级占比的饼图。

```python
l = df["等级"].value_counts()
x,y = l.index,l.values
y = y/sum(y)
plt.figure(figsize = (5,5))
plt.pie(y,labels=x)
```

（6）画出酒店评分的直方图。

```python
plt.figure(figsize = (10,6))
plt.hist(df["评分"], bins = 20, edgecolor = 'k',alpha = 0.5)
plt.show()
```

（7）画出每个热门等级酒店评分均值的柱状图。（按照评分均值从小到大排序。）
