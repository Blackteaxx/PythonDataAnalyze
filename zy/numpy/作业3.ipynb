## （2）数据操作

（1）读取香港酒店数据。

```python
df1 = pd.read_excel('香港酒店数据.xlsx')
df1.head()
```

（2）按照数据的内容，重新设置数据的索引，重新设置列名称为'名字','类型','城市','地区','地点','评分','评分人数','价格'。

```python
df1 = df1.drop('Unnamed: 0',axis=1)
df1.head()
```

```python
df1.columns = ['名字','类型','城市','地区','地点','评分','评分人数','价格']
df1.head()
```

```python
df1.index = range(1,len(df1)+1)
df1.head()
```

（3）查看所有类型为“浪漫情侣”的酒店

```python
df1[df1.类型 == '浪漫情侣']
```

（4）查看所有类型为“浪漫情侣”，地区在湾仔的酒店

```python
df1[(df1.类型 == '浪漫情侣') &(df1.地区 == '湾仔')]
```

（5）查看所有地址在观塘或者油尖旺，评分大于4的酒店

```python
df1[((df1.地区 == '观塘') |(df1.地区 == '油尖旺'))&(df1.评分>4)]
```

（6）查看类型缺失的数据

```python
df1[df1['类型'].isnull()]
```

（7）用“其他”填充类型和地区

```python
df1['类型'].fillna('其他',inplace=True)
df1['地区'].fillna('其他',inplace=True)
df1[df1['类型'].isnull()]
```

（8）用评分均值填充缺失值

```python
df1['评分'].fillna(np.mean(df1['评分']),inplace=True)
df1[:10]
```

（9）删除价格和评分人数的缺失值

```python
df1 = df1.drop(df1[df1['价格'].isnull()].index)
df1 = df1.drop(df1[df1['评分人数'].isnull()].index)
df1.head()
```

（10）保存到“酒店数据1.xlsx”

```python
df1.to_excel("香港酒店数据c.xlsx")
```
