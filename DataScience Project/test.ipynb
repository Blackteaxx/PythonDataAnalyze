```python
import pandas as pd
df = pd.read_csv("News.csv")
# df.insert(2,column="author",value=pd.NA)

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:65.0) Gecko/20100101 Firefox/65.0'}  # ä¼ªè£…å¤´
import requests,time,json
from bs4 import BeautifulSoup
def getTitleAndContent(url):
    html = requests.get(url,headers=headers).text
    soup = BeautifulSoup(html,'html.parser')
    title = soup.find("h1").text
    content = ''
    contentList = soup.find_all("p")
    for paragraph in contentList:
        content = content+paragraph.text+'\n'
    return title,content
from lxml import etree
def getAuthor(url):
    html = requests.get(url,headers=headers).text
    # soup = BeautifulSoup(html,'html.parser')
    tree = etree.HTML(html)
    author = tree.xpath("/html/body/header/div[3]/div[1]/a")[0]
    return author.text
data = {}
index = 0
urllist = df['url'].values
while index<=39643:
    try:
        author = getAuthor(url=urllist[index])
        data[urllist[index]] = author
        print(index)
        if index%1000==0:
            f = open("authorData.json","w",encoding='utf-8')
            json.dump(data,f)
            f.close()
        index+=1
    except ConnectionResetError:
        f = open("authorData.json","w",encoding='utf-8')
        json.dump(data,f)
        f.close()
        time.sleep(120)
    except:
        f = open("authorData.json","w",encoding='utf-8')
        json.dump(data,f)
        f.close()
        time.sleep(120)
f = open("authorData.json","w",encoding='utf-8')
json.dump(data,f)
f.close()
```
